```python
import pandas as pd
import pubchempy as pcp
```

```python
def get_iupac(smiles):
    try:
        # Look up compound by SMILES
        compounds = pcp.get_compounds(smiles, namespace='smiles')
        if compounds:
            return compounds[0].iupac_name
        return "Not Found"
    except Exception as e:
        return f"Error: {e}"
```

```python
# --- Configuration ---
input_csv = 'input_smiles.csv'  # Change this to your source file
output_csv = 'output_iupac.csv'
smiles_column = 'smiles'       # The name of the column containing SMILES
# ---------------------

# Create a dummy CSV if it doesn't exist for demonstration
try:
    df = pd.read_csv(input_csv)
except FileNotFoundError:
    print(f"File {input_csv} not found. Creating a sample file...")
    data = {smiles_column: ['CCO', 'C1=CC=CC=C1', 'CC(=O)O']}
    df = pd.DataFrame(data)
    df.to_csv(input_csv, index=False)

print("Translating SMILES to IUPAC... (this may take time due to API limits)")

# Apply the translation
# Note: PubChem has rate limits. For large files, consider adding time.sleep()
df['iupac'] = df[smiles_column].apply(get_iupac)

# Save result
df.to_csv(output_csv, index=False)
print(f"Results saved to {output_csv}")

# Display preview
print(df['iupac'].head())
```
